```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "data_iafq_firearms_trafficking.xlsx"  
df = pd.read_excel(file_path, engine='openpyxl')

df.columns = df.iloc[1]
df = df[2:] 

df = df[['Region', 'VALUE']].dropna()

df['VALUE'] = pd.to_numeric(df['VALUE'], errors='coerce')

data = df.groupby("Region")["VALUE"].sum()

fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    data, labels=data.index, autopct='%1.1f%%', startangle=90, 
    wedgeprops={'edgecolor': 'white'}
)
ax.add_artist(plt.Circle((0, 0), 0.5, fc='white'))  

text_x = 1.3  
text_y = 0.8  
line_spacing = 0.1  
plt.text(text_x, text_y + 0.1, "Total par Rgion :", fontsize=12, fontweight='bold', ha='left')
for region, value in data.items():
    text_y -= line_spacing
    plt.text(text_x, text_y, f"{region}: {value:,.0f}", fontsize=10, ha='left')

plt.title("saisies des armes par rgion")

plt.show()
```

Ce graphique montre la rpartition des saisies darmes par rgion. Les Amriques enregistrent la plus grande quantit darmes saisies, avec des circuits de trafic reliant divers pays du continent. LAsie suit, avec une circulation importante darmes lie  des conflits et au commerce illgal. LEurope est galement concerne, bien que les rgulations y soient plus strictes. LAfrique et lOcanie prsentent des volumes de saisies plus faibles, lAfrique tant aussi une zone de transit pour le trafic.
